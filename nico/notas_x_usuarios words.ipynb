# Librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import nltk
```

```python
# Descargar stopwords si es necesario
nltk.download('stopwords')
from nltk.corpus import stopwords as nltk_stopwords
```

```python
fileName= "notas_x_usuario.csv"
path="../franco/"

# Read the CSV file
df = pd.read_csv(path + fileName)
```

Filtro por usuario

```python
# 2. Filtrar por id_usuario
user_id = 123  # Reemplaza por el id de usuario que desees explorar
df_user = df[df['id_usuario'] == user_id]
```

```python
# 3. Concatenar todos los textos de 'contenido'
text = " ".join(df['contenido'].dropna().astype(str).tolist())
```

# Definir stopwords

```python
wc_stopwords = set(STOPWORDS)
wc_stopwords |= set(nltk_stopwords.words('english'))
# Añade stopwords específicas si las hay
custom_stopwords = {'http', 'https', 'rt','first','one','two','four','five','without','use','set'}
wc_stopwords |= custom_stopwords
```

# Generar WordCloud

```python
wc = WordCloud(
    width=800, 
    height=400, 
    background_color='white',
    stopwords=wc_stopwords,
    collocations=False
).generate(text)

```

```python
# 6. Mostrar el gráfico
plt.figure(figsize=(15, 7.5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title(f'WordCloud General', fontsize=20)
#plt.title(f'WordCloud para usuario {user_id}', fontsize=20)
plt.show()
```
